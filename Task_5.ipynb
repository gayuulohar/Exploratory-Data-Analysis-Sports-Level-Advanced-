```python
season_data=match_data[['id','season','winner']]
season_data
```

```python
complete_data=deli_data.merge(season_data,how='inner',left_on='match_id',right_on='id')
complete_data
```

# Now, check for number of matches Won by Team in each season.

```python
wins_per_season=match_data.groupby('season')['winner'].value_counts()
wins_per_season
```

```python
plt.figure(figsize = (15,10))
sns.countplot(x='season',data=match_data,palette="cool")
plt.title("Number of Matches played in each IPL season",fontsize=20)
plt.xlabel("season",fontsize=15)
plt.ylabel('Matches',fontsize=15)
plt.show()
```

Thus, here we can observe that Highest Matches played in year 2013 followed by 2012 and 2011.

# Plotting number of matches won by team.

```python
plt.figure(figsize=(15,10))
sns.countplot(x='winner',data=match_data,palette='summer')
plt.title("Number of matches won by a team",fontsize=20)
plt.xlabel("Teams",fontsize=20)
plt.ylabel("No. of wins",fontsize=20)
plt.xticks(rotation=50)
plt.show()
```

Here, from this graph we can observe that **Mumbai Indians Won Highest match followed by Chennai Super Kings.**

# Now, we plot pie chart to see percentage of match won by batting and bowlling decision.

```python
match_data['win_by']=np.where(match_data['win_by_runs']>0,'Bat First','Bowl First')
```

```python
Win=match_data.win_by.value_counts()
labels=np.array(Win.index)
sizes=Win.values
colors=['yellow','purple']
plt.figure(figsize=(10,8))
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True,startangle=90)
plt.title("Match Result",fontsize=15)
plt.axis('equal')
plt.show()
```

Thus, **Chance of winning is higher for Decision Bowl First.**

# Plotting multiple bar graph on number of matches won by batting and bowlling first.

```python
plt.figure(figsize=(14,9))
sns.countplot(x='season',hue='win_by',data=match_data,palette='Blues')
plt.title('Number of matches won by batting and bowling first',fontsize=20)
plt.xlabel("Season")
plt.ylabel("count")
plt.show()
```

# Plotting Pie chart for toss decision.

```python
Toss=match_data.toss_decision.value_counts()
labels=np.array(Toss.index)
sizes=Toss.values
colors=['pink','grey']
plt.figure(figsize=(10,8))
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True,startangle=90)
plt.title("Toss Result",fontsize=25)
plt.axis('equal')
plt.show()
```

Here, **from this there 61.2% chance to take bowl first if captain win the toss.**

# Plotting Multiple Bar graph on Numbers of matches won by Toss result.

```python
plt.figure(figsize=(15,10))
sns.countplot(x='season',hue='toss_decision',data=match_data,palette='hsv')
plt.title("Number of matches won by toss result",fontsize=20)
plt.xlabel("Season",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

# Print numbers of fours hit by team


```python
four_data=complete_data[complete_data['batsman_runs']==4]
four_data.groupby('batting_team')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')])
```

# Print no of sixes hit by team

```python
six_data=complete_data[complete_data['batsman_runs']==6]
six_data.groupby('batting_team')['batsman_runs'].agg([('runs by six','sum'),('sixes','count')])
```

Thus, from this we can say that **Maximum Fours(4) and sixes(6) were hit by Chennai Super Kings.** 

# Print ipl most wicket taking bowlers.

```python
wicket_data=deli_data.dropna(subset=['dismissal_kind'])
wicket_data=wicket_data[~wicket_data['dismissal_kind'].isin(['run out','retired hurt','obstructing the field'])]
```

```python
wicket_data.groupby('bowler')['dismissal_kind'].agg(['count']).reset_index().sort_values('count',ascending=False).reset_index(drop=True).iloc[:10,:]
```
